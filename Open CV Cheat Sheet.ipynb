# Reshaping and Resizing Image

Reshaping and resizing images are great ways to space and computational time and is sometime required by some deep learning model especially in transfer learning.

Let's write a function to resize images this way we can use it in images as well as videos

```python
def resize_image(image, scale=0.75):
    ''' Function to resize a image or a frame'''
    
    width = (int) (image.shape[1] * scale) # rescaling width
    height = (int) (image.shape[0] * scale) # rescaling height
    dimensions = (width, height) # dimension tuple to find in resize function
    return cv.resize(image, dimensions)
```

```python

image = cv.imread('Images/art1.png') # importing images
resized_image = resize_image(image, scale=0.5)
# Correcting color channels for matplotlib
resized_image = cv.cvtColor(resized_image, cv.COLOR_BGR2RGB)

fig, ax1 = plt.subplots(ncols=2, figsize=(10, 5))
# Displaying original Image
ax1[0].imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
ax1[0].set_title('Original Image')

# Displaying Resized Image
ax1[1].imshow(resized_image)
ax1[1].set_title('50% Resized Image')

plt.show()
```

As we can see on scale image on right is 50% smaller than image in left.

Let's see how rescaling works on videos.

```python
vid_capture = cv.VideoCapture('videos/vid2.mp4')
while vid_capture.isOpened():
    ret, frame = vid_capture.read()
    
    if not ret:
        print('Frame not available')
        break
    
    frame_resize = resize_image(frame, scale=0.5) # Resizing frame
    
    cv.imshow('Original Video', frame) # Displaying original video
    cv.imshow('Resized video', frame_resize) # Resized video
    
    if cv.waitKey(1) == ord('q'):
        break
vid_capture.release()
cv.destroyAllWindows()
```

# Important Function

Gaussian Blur, Blur allows us to decrease noise of an image there are many blurring algorithm but Gaussian blur is most commonly used. Gaussian blur takes kernel size as an input which is a hyper-parameter to blur size.

Then we will detect edges in an image using canny edge detector and we will increase or decrease edge size using dilate and erode functions. At last we will rescale and crop image

```python
def bgr_to_rgb(image):
    ''' Function to convert BGR image to RGB image'''
    return cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
# Importing image
original_image = cv.imread('Images/scene.png')

# Subplots
fig, ax2 = plt.subplots(nrows=2, ncols=3, figsize=(15, 5))

# Displaying Original Image
ax2[0,0].imshow(bgr_to_rgb(original_image))
ax2[0,0].set_title('Original Image')


# Guassian Blur
kernel_size = (7,7)
blured_image = cv.GaussianBlur(original_image, kernel_size, cv.BORDER_DEFAULT)
# Displaying blurred image
ax2[0,1].imshow(bgr_to_rgb(blured_image))
ax2[0,1].set_title('Blurred Image')

# Edge Cascade
canny = cv.Canny(blured_image, 100, 165)
# Displaying Edges
ax2[0,2].imshow(bgr_to_rgb(canny))
ax2[0,2].set_title('Canny Image')

# Dilated Image
kernel_size = (3,3)
dilate = cv.dilate(canny, kernel_size, iterations=1)
# Displaying Dilated edges
ax2[1,0].imshow(bgr_to_rgb(dilate))
ax2[1,0].set_title('Dilated Image')

# Eroding Image
erode = cv.erode(dilate, kernel_size, iterations=2)
# Displaying Eroded edges
ax2[1,1].imshow(bgr_to_rgb(erode))
ax2[1,1].set_title('Eroded Image')

# Resizing Image
resize_size = (250, 250)
resize_image = cv.resize(original_image, resize_size, interpolation=cv.INTER_AREA)
# Displaying resized Images
ax2[1,2].imshow(bgr_to_rgb(resize_image))
ax2[1,2].set_title('Resized Image')

plt.show()
```

```python
# Cropping Image
cropped_image = original_image[0:250, 250:550]
plt.imshow(bgr_to_rgb(cropped_image))
plt.title('Cropped Image')
plt.show()
```

# Image Transformations

Common techniques which are applied to images such as translation, rotation and flipping

## Translation

```python
image = cv.imread('Images/art2.png')
# Translation, shifiting image up, down, left, right or any combination of these operations

def translation(image, x, y):
    '''Function to shift image
        +X --> Shift right
        -X --> Shift left
        +y --> Shift down
        -y --> Shift up
    '''
    translate_Mat = np.float32([[1,0, x], [0, 1, y]]) # Shifted matrix
    dimensions = (image.shape[1], image.shape[0]) # Dimensions of image (Width, Height)
    
    return cv.warpAffine(image, translate_Mat, dimensions)


tranlated_image = translation(image, 500, 1000)
fig, ax4 = plt.subplots(ncols=2, figsize=(10, 5))
ax4[0].imshow(bgr_to_rgb(image))
ax4[0].set_title('Original Image')

ax4[1].imshow(bgr_to_rgb(tranlated_image))
ax4[1].set_title('Translated Image')

plt.show()
    
```

## Rotation

```python
image = cv.imread('Images/art1.png')

# Image can be rotated on any pivot point in opencv
def rotation(image, angle, pivot=None):
    '''Function to rotate an Image
        + angle --> anticlockwise rotation
        - angle --> clockwise rotation'''
    (width, height) = image.shape[1], image.shape[0] # Extracting width and height of image
    
    if pivot is None:
        pivot = (width//2, height//2)
    
    rotMat = cv.getRotationMatrix2D(pivot, angle, 1.0)
    dimension = (width, height)
    
    return cv.warpAffine(image, rotMat, dimension)


fig, ax5 = plt.subplots(ncols=2, figsize=(10, 5))
ax5[0].imshow(bgr_to_rgb(image))
ax5[0].set_title('Original Image')

rotated_image = rotation(image, -30)
ax5[1].imshow(bgr_to_rgb(rotated_image))
ax5[1].set_title('Roated Image')
plt.show()
```

# Contour Detection

# Color Spaces

How to split and merge color channels, we are going to split image in its 3 color channels

```python
image = cv.imread('Images/scene.png')

# Spliting in bgr color channels
b,g,r = cv.split(image)
print('Original shape', image.shape)
print('Shape of B, G, R =', b.shape, g.shape, r.shape)

cv.imshow('Blue', b)
cv.imshow('Green', g)
cv.imshow('Red', r)

cv.waitKey(0)
```

This shows pixel's intensity for each color channel and each of these color channels defines how much that color is in specific pixel. Now lets merge these channels again.

```python
merged_image = cv.merge([b, g, r])
plt.imshow(bgr_to_rgb(merged_image))
plt.title('Merged Image')
plt.show()
```

Its time look at each color channel not in grayscale but in their respective color channel to do this we need to make 1 channel image 3 channel using two blank channels.

```python
# Creating blank channels
blank = np.zeros(image.shape[:2], dtype='uint8')

# Generating 3 channels of all colors
blue_image = cv.merge([b, blank, blank])
green_image = cv.merge([blank, b, blank])
red_image = cv.merge([blank, blank, r])

# Displaying image using matplotlib
fig, ax11 = plt.subplots(ncols=3, figsize=(15,5))
ax11[0].imshow(bgr_to_rgb(blue_image))
ax11[0].set_title('Blue Image')

ax11[1].imshow(bgr_to_rgb(green_image))
ax11[1].set_title('Green Image')

ax11[2].imshow(bgr_to_rgb(red_image))
ax11[2].set_title('Red Image')

plt.show()
```

# Blurring

How to smooth and blur and image, we generally do this to discard noise from the image by applying blur. Gaussian blur is most popular but not suitable everytime.

Blurring use concept of convolution on an image. Convolution mathematically means in context of convolving two functions. Here we can this of image as an function and blur filter as another function, blur filter function is called kernel.


During convolve function we march kernel over our image and multiply overlapping pixels and add all elements of kernel. Higher the kernel size more blur there is.

```python
image = cv.imread('Images/us.png')
kernel_size = (7,7)
plt.imshow(bgr_to_rgb(image))
plt.title('Original Image')
plt.show()
```

## Averaging Blur

This kernel takes average of all pixels.

```python
avg = cv.blur(image, kernel_size)
plt.imshow(bgr_to_rgb(avg))
plt.title('Avergae blurring')
plt.show()
```

## Gaussian Blur

It uses normal distribution as compared to average blur therefore it requires standard deviation as a hyper-parameter

```python
standard_deviation = 1.0
gauss = cv.GaussianBlur(image, kernel_size, standard_deviation)
plt.imshow(bgr_to_rgb(gauss))
plt.title('Gaussian blur')
plt.show()
```

## Median Blur

Similar to average blurring but it calculates median. It is more effective than average blurring in noise reduction hence used in advanced computer vision projects

```python
median = cv.medianBlur(image, 7)
plt.imshow(bgr_to_rgb(median))
plt.title('Median blurring')
plt.show()
```

## Bilateral Blurring

Most effective, It retains edges in images therefore more effective

```python
diameter = 50
sigmacolor = 15 # larger the values more the colors are considerd in neighbourhood
sigmaspace = 35 # larger the value more outer pixel will influence blurring
bilateral = cv.bilateralFilter(image, diameter, sigmacolor, sigmaspace)
plt.imshow(bgr_to_rgb(bilateral))
plt.title('Bilateral blurring')
plt.show()
```

# Bitwise Operators

Bitwise operators work similar as you would expect binary operations. There are 4 types of bitwise operators. And, Or, Xor and not.
These bitwise operations are useful in masking an image.

```python
# Creating a blank image to work on
image_size = (500, 500)
blank = np.zeros(image_size,dtype='uint8')

# Recatangle on a blank image
recatangle = cv.rectangle(blank.copy(), (30, 30), (470, 470), color=255, thickness=-1)

# Circle on blank image
circle = cv.circle(blank.copy(), (blank.shape[1]//2, blank.shape[0]//2), 250, color=255, thickness=-1)

# Plotting images
fig, ax12 = plt.subplots(nrows=2, ncols=3, figsize=(15,7))

ax12[0,0].imshow(bgr_to_rgb(recatangle))
ax12[0,0].set_title('Recatangle')

ax12[0,1].imshow(bgr_to_rgb(circle))
ax12[0,1].set_title('Circle')

# Bitwise and
and_image = cv.bitwise_and(recatangle, circle)
ax12[0,2].imshow(bgr_to_rgb(and_image))
ax12[0,2].set_title('AND operation')

# Bitwise or
or_image = cv.bitwise_or(recatangle, circle)
ax12[1,0].imshow(bgr_to_rgb(or_image))
ax12[1,0].set_title('OR operation')

# Bitwise xor
xor_image = cv.bitwise_xor(recatangle, circle)
ax12[1,1].imshow(bgr_to_rgb(xor_image))
ax12[1,1].set_title('XOR operation')

# Bitwise Not
not_image = cv.bitwise_not(recatangle)
ax12[1,2].imshow(bgr_to_rgb(not_image))
ax12[1,2].set_title('Not operation')

plt.show()
```

# Masking

Masking is an useful operation often used to isolate the important part of the image. One thing to note mask image must have same dimensions as the original image.

```python
image = cv.imread('Images/art1.png')

# Create a blank image of size same as image
blank = np.zeros(image.shape[:2], dtype='uint8')

# Creating a mask
mask = cv.circle(blank, (image.shape[1]//2, image.shape[1]//2-50), 250, color=255, thickness=-1)

masked_image = cv.bitwise_and(image, image, mask=mask)

plt.imshow(bgr_to_rgb(masked_image))
```

# Histogram Computation

Here we compute histogram containing pixel intensity of image divided in bins. This part is divided into two parts. Grayscale and colored

## Grayscaled

```python
image = cv.imread('Images/scene.png')

# Gray scaling image
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Calculating histogram without mask
hist = cv.calcHist([gray], [0], mask=None, histSize=[256], ranges=[0, 256])

# Calculating hist with mask
blank = np.zeros(gray.shape[:2], dtype='uint8')
circle = cv.circle(blank, (gray.shape[1]//2, gray.shape[0]//2), 100,color=255, thickness=-1)
mask = cv.bitwise_and(gray, gray, circle)
hist_mask = cv.calcHist([gray], [0], mask= mask, histSize=[256], ranges=[0, 256])

# Plotting 
fig, ax13 = plt.subplots(nrows=3, figsize=(7, 10))
ax13[0].imshow(bgr_to_rgb(gray))
ax13[0].set_title('Gray Image')

ax13[1].plot(hist)
ax13[1].set_title('Hist without mask')

ax13[1].plot(hist)
ax13[1].set_title('Hist without mask')

ax13[2].plot(hist_mask)
ax13[2].set_title('Hist with mask')

plt.show()
```

## Colored

Calculating histogram for colored image is mildly different we calculate histogram for each color channel

```python
color = ('b', 'g', 'r')

for i, color in enumerate(color):
    hist = cv.calcHist([image], [i], None, [256], [0, 256])
    
    plt.plot(hist, color=color)
    plt.xlim([0,256])
    plt.ylim([0, 10000])

plt.xlabel('Bins')
plt.ylabel('No. of pixels')
plt.title('Color Histogram')
plt.show()
    
```

# Edge Detection

There are two major edge detection algorithm based on computing gradients Laplacian and Sobel.

```python
image = cv.imread('Images/art3.png')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

fig, ax15 = plt.subplots(nrows=2, ncols=3, figsize=(15,5))
ax15[0,0].imshow(bgr_to_rgb(image))
ax15[0,0].set_title('Original Image')

# Laplacian gradient edges
lap = cv.Laplacian(gray, cv.CV_64F)
lap = np.uint8(np.absolute(lap))

ax15[0,1].imshow(bgr_to_rgb(lap))
ax15[0,1].set_title('Laplacian Image')

# Sobel x and y
sobel_x = cv.Sobel(gray, cv.CV_64F, 1, 0)
sobel_y = cv.Sobel(gray, cv.CV_64F, 0, 1)

ax15[0,2].imshow(sobel_x, cmap='gray', vmin=0, vmax=255)
ax15[0,2].set_title('Sobel x Image')

ax15[1,0].imshow(sobel_y, cmap='gray', vmin=0, vmax=255)
ax15[1,0].set_title('Sobel y Image')

# Sobel combined
combined_sobel = cv.bitwise_or(sobel_x, sobel_y)
ax15[1,1].imshow(combined_sobel, cmap='gray')
ax15[1,1].set_title('Sobel y Image')

plt.show()
```

# Face Detection (Haar Cascade)

There is difference between face recognition and face detection. Face detection only detect a face in image where as face recognition not only detects a face but also tells whose face it is.

face detection is usually performed by a binary classifier and open cv already contains lots of pre-trained classifiers one of those is Haar cascade. It just uses edges so, we need grayscale images. Haar cascades are not immune to noise hence it is not supposed to detect faces correctly. Deled face recognizer is more general for practical use case.

```python
# Importing image with a face
face = cv.imread('Images/poster1.png')

# Gray scaling an image
gray_face = cv.cvtColor(face, cv.COLOR_BGR2GRAY)

# Haar cascade classifier initialization
haar_cascade = cv.CascadeClassifier('haar_face.xml')

# Prediction, it return rectangular coordinate
# Increasing minNeighbours decreases sensitivity of model
faces_rect = haar_cascade.detectMultiScale(gray_face, scaleFactor=1.2, minNeighbors=5)

print('Number of faces =', len(faces_rect))

# Now we can loop over images and draw a rectangle over the images
for (x, y, w,h) in faces_rect:
    pt1 = (x,y)
    pt2 = (x+w,y+h)
    cv.rectangle(face, pt1, pt2, color=(0,0,255), thickness=10)
plt.figure(figsize=(15,5))
plt.imshow(bgr_to_rgb(face))
plt.show()
```

## Training Model

```python
import os
```

```python
# make list of all the people in images
people = []
for i in os.listdir(r'Images/Faces/train'):
    people.append(i)
print(p)
```

```python
dir = 'Images/Faces/train'
```

```python
features = []
labels = []
def create_train():
    '''To loop over each folder and images and import it '''
    for person in people:
        path = os.path.join(dir, person)
        label = people.index(person)
        
        for img in os.listdir(path):
            img_path = os.path.join(path, img)
            
            img_array = cv.imread(img_path)
            gray = cv.cvtColor(img_array, cv.COLOR_BGR2GRAY)
            
            # Lets detect faces 
            faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=2) 
            
            # Extracting face region
            for (x, y, w, h) in faces_rect:
                faces_roi = gray[y:y+h, x:x+w]
                features.append(faces_roi)
                labels.append(label)
```

```python
create_train()
print('length of features list =', len(features))
print('length of labels list =', len(labels))
```

```python
# Initializing face recognizer model
face_recognizer = cv.face.LBPHFaceRecognizer_create()

# trainging the face recoginizer on features list and labels by converting it to np array
features = np.array(features, dtype='object')
labels = np.array(labels)
face_recognizer.train(features, labels)

# Saving model and that so that we can it anywhere else
np.save('features.npy', features)
np.save('labels.npy', labels)
face_recognizer.save('face_recognizer.yml')
print('Model trained')
```
